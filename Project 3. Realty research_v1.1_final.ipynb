для данных случаев возможна замена пропусков:

 * 13  balcony *заменим на 0, то есть пропуск = нет балкона*               
 * 17  parks_around3000 *заменим на 0, то есть пропуск = нет парков в радиусе 3000 м*   
 * 19  ponds_around3000 *заменим на 0, то есть пропуск = нет парков в радиусе 3000 м*      
        

Заполняем пропуски для выбранных столбцов:

### Изменение типов данных

Проверим типы данных в столбцах датафрейма

```python
data.info()
```

Тип данных "object" применяется только для 2 столбцов: locality_name и first_day_exposition, что оправдано, так как в первом случае столбец хранит название населенного пункта, а во втором закодированную дату публикации. Дату публикации позднее нужно будет перевести в формат даты и времени.

Проверим данные, которые хранятся в столбцах

```python
for column in data.columns:
    print(data[column].value_counts()) #в цикле с помощью метода value_counts выводим списки уникальных значений по столбцам
    print('------------------------------------------------------')
```

```python
for column in data.columns:
    print(data[column].describe()) #в цикле с помощью метода describe выводим числовое описание данных
    print('-------------------------------------------------------')
```

Критичных проблем с типами данных не выявлено. Однако во многих столбцах использование типа данных float64 (вещественное число) избыточно, можно изменить тип данных на int64 (целочисленный).
Также стоит привести столбец first_day_exposition к формату даты и времени.

```python
for column in data.columns: # проходим в цикле по столбцам датафрейма и меняем тип данных с float64 на int64, где это возможно
    if data[column].dtype == 'float64':
        data[column] = data[column].astype('int64', errors = 'ignore')

data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')
# переводим столбец first_day_exposition в формат даты и времени
data.info()
```

Таким образом, мы привели часть столбцов c вещественным типом данных float64 к целочисленному типу int64.

### Расчет вспомогательных параметров

Для целей дальнейшего анализа рассчитаем ряд параметров: 

* цену квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

Добавляем цену квадратного метра, день недели, месяц, год публикации объявления

```python
data['m2_price'] = data['last_price'] / data['total_area'] # рассчитываем стоимость квадратного метра жилплощади
data['m2_price'] = data['m2_price'].round(0) # округляем результат до третьего знака после запятой
data['weekday_exp'] = pd.DatetimeIndex(data['first_day_exposition']).weekday # добавляем день недели с помощью метода DatetimeIndex
data['month_exp'] = pd.DatetimeIndex(data['first_day_exposition']).month # добавляем месяц
data['year_exp'] = pd.DatetimeIndex(data['first_day_exposition']).year # добавляем год
```

Добавляем категорию этажа квартиры, для этого создаем функцию определения этажа (первый, последний, другой)

```python
def floor_def(row):
    if row['floor'] == 1:
        return 0 # 'первый'
    elif row['floor'] == row['floors_total']:
        return 0.5 #'последний'
    else:
        return 1 # 'другой'

data['floor_category'] = data.apply(floor_def, axis = 1) # применяем созданную функцию методом apply
display(data.head(10)) # выводим и проверяем полученный результат
```

Рассчитываем соотношение жилой и общей площади, а также отношение площади кухни к общей. Также для удобства посчитаем цену недвижимости в миллионах рублей.

```python
data['living_total_area_ratio'] = data['living_area'] / data['total_area'] # рассчитываем соотношение жилой и общей площади
data['kitchen_total_area_ratio'] = data['kitchen_area'] / data['total_area'] # рассчитываем соотношение площади кухни к общей площади
data['living_total_area_ratio'] = data['living_total_area_ratio'].round(2) # округляем результат до 2 знаков после запятой
data['kitchen_total_area_ratio'] = data['kitchen_total_area_ratio'].round(2) # округляем результат до 2 знаков после запятой
data['last_price_mil'] = data['last_price'] / 1000000 # рассчитываем стоимость жилья в млн. руб.
data['last_price_mil'] = data['last_price_mil'].round(2) # округляем результат до 2 знаков после запятой

display(data.head()) # выводим результат
```

#### Выделение центральной зоны Санкт-Петербурга
