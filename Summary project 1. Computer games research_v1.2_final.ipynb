## Предобработка данных

### Корректировка названий столбцов

Сначала приведем названия столбцов к нижнему регистру

```python
data.columns = data.columns.str.lower() #приводим названия столбцов датафрейма к нижнему регистру методом str.lower
data.info() # проверяем результат
```

### Преобразование типов данных

Разберемся с типами данных. Год релиза (year_of_release) хранится в переменной с вещественным типом данных. Преобразуем его в целочисленный. Аналогичным образом поступим со столбцом critic_score, который хранит оценку критиков. Также данные с типом float64 преобразуем в тип float32 для экономии памяти.

Отдельно стоит отметить, что у столбца user_score тип данных object, хотя в нем должны храниться только числовые данные. Необходимо будет проверить данный столбец отдельно.

```python
data['year_of_release'] = data['year_of_release'].astype('Int32')
data['critic_score'] = data['critic_score'].astype('Int32')
# преобразуем столбцы в тип данных Int32 (поддерживает NA значения)


for column in data.columns: # проходим в цикле по столбцам датафрейма и меняем тип данных с float64 на float32, где это возможно
    if data[column].dtype == 'float64':
        data[column] = data[column].astype('float32')

display(data.head()) # проверяем результат
display(data.info())
```

### Поиск и обработка пропусков данных

Проверим данные на пропуски и при необходимости заполним пропущенные значения. Посчитаем долю пропусков в каждом столбце.

```python
data.isna().sum() / data['platform'].count()
```

Доля пропусков в данных достигает 51%. Основные столбцы с пропусками данных: critic_score, user_score, rating. Данные столбцы содержат в себе информацию об оценке критиков, пользователей, а также о рейтинге от организации ESRB. Заполнять пропущенные значениями искусственно было бы некорректно, так как у разных игр может быть совершенно разная оценка пользователей, даже в рамках одного жанра или платформы.

Кроме того, по 2 пропуска присутствуют в столбцах name и genre, а также для 2% строк отсутствуют данные по году релиза (year_of_release). Учитывая небольшой объем пропусков, ими можно пренебречь. Заполнять данные искусственными значениями в данном случае некорректно. Удалим такие строки из датасета.

```python
data = data.dropna(subset = ['name', 'year_of_release', 'genre']) # удаляем строки с пропущенными значениями методомо dropna
data['year_of_release'] = data['year_of_release'].astype('int32') 
#после удаления na значений преобразовываем тип данных в столбце year_of_release в int32 
#(в дальнейшем мы будем использовать метод query. Тип int32 поддерживается методом query, в отличие от Int32)
data.info() 
data.isna().sum() / data['platform'].count()
# проверяем результат
```

Проанализируем подробнее столбцы с большой долей пропусков в данных. Выведем количество и долю пропусков по годам.

```python
data['critic_score_na'] = data['critic_score'].isna() 
data['user_score_na'] = data['user_score'].isna()
data['rating_na'] = data['rating'].isna()
# создаем столбцы с подсчетом пропусков в данных
display(data.head()) # выводим получившийся массив данных

data_pivot_na = (data.pivot_table(
    index = 'year_of_release', 
    values = ['critic_score_na', 'user_score_na', 'rating_na','platform'], 
    aggfunc = {'critic_score_na':'sum','user_score_na':'sum','rating_na':'sum','platform':'count'}))

# строим сводную таблицу методом pivot_table по годам

data_pivot_na['critic_score_na_share'] = data_pivot_na['critic_score_na'] / data_pivot_na['platform']
data_pivot_na['user_score_na_share'] = data_pivot_na['user_score_na'] / data_pivot_na['platform']
data_pivot_na['rating_na_share'] = data_pivot_na['rating_na'] / data_pivot_na['platform']

# считаем долю пропусков в каждом столбце по годам

display(data_pivot_na) # выводим итоговую таблицу с данными
```

Как видно из получившейся таблицы, пропуски присутствуют в каждом срезе по году, в некоторых годах их доля достигает 100%. Скорее всего причина в том, что не для всех игр доступна официальная оценка пользователей и критиков, а также рейтинг от ESRB.

Так как рейтинг ESRB используется только для маркировки игр для США и Канады, то удалять пропущенные значения в данном столбце точно не следует, оставим пропущенные значения и установим на них "заглушку".

```python
data['rating'] = data['rating'].fillna('no_esrb')
```

Также отдельно проанализируем столбец с оценкой пользователей user_score. Тип данных данного столбца object, хотя храниться там должны только числовые данные.

```python
data['user_score'].value_counts()
```

Наиболее часто встречается значение 'tbd', что означает 'to be defined' - ожидает оценки. Для удобства дальнейших расчетов заполним его пропусками и переконвертируем столбец в числовой формат.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors = 'coerce', downcast = 'integer') 
# преобразовываем данные в числовой формат методом to_numeric с парметром errors=coerce, который заменит 'tbd' на пропуски значений 
data['user_score'] = data['user_score'].astype('float32')
data.info() # проверяем результат 
```

Наконец, проверим данные на дубликаты.

```python
data.duplicated().sum()
```

Дубликатов не обнаружено, переходим к следующему этапу.

Рассчитаем суммарные продажи игр по всем регионам

**Выводы:** Предобработка данных завершена. Мы привели названия столбцов к нижнему регистру, преобразовали данные в нужные типы, а также выявили и объясняли пропуски в данных. Помимо этого были рассчитаны суммарные продажи каждой игры по всем регионам.

### Распределение платформ по годам

Проверим, как распределяются продажи по платформам и как меняется это распределение со временем.

Определим основные игровые платформы по объему продаж.

```python
popular_platforms = data.pivot_table(index = 'platform', values = 'ttl_sales', aggfunc = 'sum')
# группирум данные по платформам и считаем суммарный объем продаж с помощью сводной таблицы
popular_platforms = popular_platforms.sort_values(by = 'ttl_sales', ascending = False) #сортируем данные по убыванию
popular_platforms = popular_platforms.reset_index() # сбрасываем индексы
popular_platforms #проверяем результат
```

Всего в данных есть информация о 31 платформе, лидеры рейтинга Play Station 2, Xbox 360 и Play Station 3. Выберем платформы с суммарным объемом продаж более 500 млн. копий

```python
popular_platforms = popular_platforms.query('ttl_sales > 255') # создаем срез данных
popular_platforms_list = popular_platforms['platform'].values.tolist() # преобразуем в список самых популярных платформ
print(popular_platforms_list) #проверяем результат
```

```python
data_popular_platforms = data.query('platform in @popular_platforms_list and year_of_release > 1990')
#создаем срез данных для анализа по ранее созданному списку самых популярных платформ
```

```python
# строим гистограмму, которая показывает количество игр, выпущенных на самых популярных платформах по годам
sns.histplot(
    data_popular_platforms,
    x='year_of_release', hue="platform",
    multiple='stack',
    #palette='light:m_r',
    edgecolor='.3',
    linewidth=.5,
    #log_scale=True,
)
```

По графику видно, что в среднем срок жизни платформы составляет около 10 лет. При этом игровые тренды меняются гораздо быстрее, поэтому возьмем для анализа период в 4 года с 2013 по 2016 г. Несмотря на то, что данный период характеризуется значительным снижением количества выпускаемых игр, он является наиболее актуальным для анализа.

### Взаимосвязь объема продаж и оценки пользователей и критиков

Проверим, есть ли зависимость между объемом продаж и оценкой пользователей и критиков компьютерных игр. Выберем одну популярную платформу для проведения исследования, например, PS4.

Построим матрицы корреляции и графики рассеяния.

```python
print('-----------PS4------------')

PS4 = data_actual.query('platform == "PS4"') # делаем срез по платформе PS4
PS4 = PS4.dropna() # удаляем строки с пропусками, так как в данном случае пропуски в оценках критиков и пользователей критичны

display(PS4[['ttl_sales', 'critic_score', 'user_score']].corr().style.background_gradient())
# строим  матрицу коэффициентов корреляции Пирсона

sns.set(rc={'figure.figsize':(20,10)}) # устанавливаем размер графика
sns.set(font_scale = 1.1) # устанавливаем размер шрифта

sns.scatterplot(data = PS4, x = 'critic_score', y = 'ttl_sales', size = 'user_score', hue = 'user_score', sizes = (20,200), alpha = 0.7)
# строим график рассеяния
```

Согласно матрице коэффициентов коррелляции между оценкой критиков и объемом продаж существует средняя положительная зависимость. В то же время, между оценкой пользователей и объемом продаж как таковая зависимость не прослеживается. Дополнительно стоит отметить, что оценки пользователей и критиков хорошо согласуются друг с другом, между ними существует сильная положительная зависимость.

Исходя из графика рассеяния можно сделать вывод, что большая часть продаж игр приходится на диапазон оценки критиков "выше среднего". Порог в 2 млн. проданных копий преодолевают игры только с оценкой критиков выше 70 баллов из 100 возможных. С оценкой пользователей не все так однозначно. Высокие оценки пользователей (от 6 из 10 баллов) встречаются у игр с низкими продажами.

Проверим данное наблюдение по оценкам пользователей.

```python
sns.scatterplot(data = PS4, x = 'user_score', y = 'ttl_sales', alpha = 0.7)
# строим график рассеяния
```

Оценки пользователей довольно сильно расходятся с объемами продаж. Так, зачастую высокие оценки получают игры с низкими объемами продаж. С лругой стороны, отдельные игры с высокими объемами продаж (от 2 млн. проданных копий) получают низкие оценки пользователей (менее 5 баллов).

На примере самой популярной платформы PS4 между оценками критиков и объемами продаж существует сильная положительная зависимость: большого объема продаж, как правило, достигают игры с высокой оценкой критиков. 

В то же время, между оценкой пользователей и объемом продаж связи не установлено. Высокий объем продаж может достигаться играми как с низкой, так и высокой оценкой пользователей, хотя, конечно, количество бестселлеров с высокой оценкой пользователей больше.

Статистически связь между оценкой пользователей и объемом продаж прослеживается только на отдельных платформах (WiiU), между оценкой критиков и объемом продаж существует средняя положительная связь.

## Проверка гипотез

Проверим следующие гипотезы:

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

### Пользовательские рейтинги платформ Xbox One и PC

Сформулируем нулевую и альтернативную гипотезы:

**Гипотеза H0:** Средние пользовательские рейтинги платформ Xbox One и PC равны

**Гипотеза H1:** Средние пользовательские рейтинги платформ Xbox One и PC различаются

**Методика проверки гипотез:**

Для проверки гипотез будем использовать t-test и специальный метод Python для проверки гипотезы о равенстве средних двух генеральных совокупностей по взятым из них выборкам

Пороговое значение alpha (уровень значимости) зададим как 0.05 (5%)

Перед использованием метода проверим, соответствуют ли выборки условиям использования t-test:

* Генеральные совокупности не должны зависеть друг от друга.
* Выборочные средние должны быть нормально распределены.
* Дисперсии рассматриваемых генеральных совокупностей должны быть равны.

Генеральные совокупности не зависят друг от друга, так как пользователи игр проставляют свои оценки независимо. Выборочные средние нормально распределены благодаря ЦПТ. У нас нет уверенности в равенстве дисперсий рассматриваемых генеральных совокупностей, поэтому для целей анализа установим параметр equal_var = False.

Для начала оценим средние рейтинги платформ

```python
XOne = data_actual.query('platform == "XOne"')
XOne = XOne.dropna() # отбрасываем пропущенные значения, так как для нас важны строки с заполненным рейтингом пользователей
print('XOne', XOne['user_score'].mean())

PC = data_actual.query('platform == "PC"')
PC = PC.dropna() # отбрасываем пропущенные значения, так как для нас важны строки с заполненным рейтингом пользователей
print('PC', PC['user_score'].mean())
```

Подтвердим полученные результаты статистически

```python
from scipy import stats as st # импортируем библиотеку stats
alpha = 0.05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(PC['user_score'],XOne['user_score'], equal_var = False)
# используем метод ttest для средних генеральных совокупностей
print('p-значение', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Не получилось отвергнуть нулевую гипотезу, средние пользовательские рейтинги XBox One и PC равны. 

### Пользовательские рейтинги жанров Action и Sports

Проверим, различаются ли средние пользовательские рейтинги игр жанров Action и Sports. Сформулируем нулевую и альтернативную гипотезы:

**Гипотеза H0:** Средние пользовательские рейтинги жанров Action и Sports равны

**Гипотеза H1:** Средние пользовательские рейтинги жанров Action и Sports различаются

Для начала оценим средний рейтинги жанров

```python
action = data_actual.query('genre == "Action"')
action = action.dropna() # отбрасываем пропущенные значения, так как для нас важны строки с заполненным рейтингом пользователей
print('action', action['user_score'].mean())

sports = data_actual.query('genre == "Sports"')
sports = sports.dropna() # отбрасываем пропущенные значения, так как для нас важны строки с заполненным рейтингом пользователей
print('sports', sports['user_score'].mean())
```

Подтвердим полученные результаты статистически

```python
alpha = 0.05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(action['user_score'],sports['user_score'], equal_var = False)
# используем метод ttest для средних генеральных совокупностей
print('p-значение', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод:** Результат показывает, что нужно отвергнуть нулевую гипотезу, это означает, что средние пользовательские рейтинги жанром action и sports значимо различаются.
